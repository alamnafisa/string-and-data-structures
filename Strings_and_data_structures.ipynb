```python
#Define a function called isSymmetrical that takes a string value as a parameter and returns TRUE if the given string 
#is Symmetrical. A string is said to be symmetrical if the reverse of the string is the same as string. 
#For example, “radar” is symmetrical, but “rehab” is not symmetrical.

def isSymmetrical(x):
    lastLetter=""
    strLen= len (x)
    indexDecr=1
    reverseWord=""
    for i in x:
        lastLetter= x[strLen-indexDecr]
        reverseWord= reverseWord + lastLetter
        #print (reverseWord)
        indexDecr= indexDecr+1
    print (reverseWord)        
    if reverseWord== x:
        result= "is a symmetrical string"
        #print ("The string you entered", '"'+ x+ '"' , "is a symmetrical string")
        return (x, result, True)
    else:
        #print ("The string you entered", '"'+x+'"', "is not a symmetrical string")
        result= "is not a symmetrical string"
        return (x, result, False)

inputDat= input( "Enter a string value you would like to process: \t" )
startProgram1= isSymmetrical (inputDat)
#print (startProgram1)
print ("The string you entered", '"'+startProgram1[0]+'"', startProgram1[1], "therefore result is", startProgram1[2])
```
